```python
#Importando bibliotecas a serem usadas
import pandas as pd
from datetime import datetime

#Baixando arquivos correspondentes no formato .csv
df = pd.read_csv('graficos.csv')
# Standard plotly imports
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import iplot, init_notebook_mode
# Using plotly + cufflinks in offline mode
import cufflinks
cufflinks.go_offline(connected=True)
init_notebook_mode(connected=True)
```

```python
df
```

```python
data = [
    go.Scatter(
        x=df['IPT_x'],
        y=df['ipt_campanha_x'],
        text=df['id_usuario'],
        mode='markers',
        marker=dict(
            sizemin=10,
            size=df['numero_ultimas_compras_x'],
            colorscale='Rainbow',
            showscale=True,
            color=df['delta_ipt'],
            line=dict(color='black', width=1.2)))
]

figure = go.Figure(
    data=data,
    layout=go.Layout(
        xaxis=dict(title='IPT Médio do Cliente'),
        yaxis=dict(title='IPT da Campanha (Novo Ciclo)'),
        title='Reads vs Word Count Colored by Read Time and Sized by Read Ratio'
    ))

iplot(figure)
```
